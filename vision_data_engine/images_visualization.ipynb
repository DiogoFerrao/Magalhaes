```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Paths to images
image_paths = [
    "/media/magalhaes/schreder/images/d1_21163_16738.jpg",
    "/media/magalhaes/coco/images/val2017/000000138979.jpg",
    "/media/magalhaes/DAWN/Fog/images/foggy-090.jpg",
    "/media/magalhaes/DAWN/Rain/images/rain_storm-054.jpg",
    "/media/magalhaes/emergency_vehicles/images/d911_152.jpg",
    "/media/magalhaes/ExDark/images/2015_02187.jpg"
]

# Name of the images
image_names = [
    "Schréder",
    "COCO",
    "DAWN Fog",
    "DAWN Rain",
    "Emergency Vehicles",
    "ExDark"
]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for i, ax in enumerate(axes.flat):
    image = plt.imread(image_paths[i])
    ax.imshow(image)
    
    # Show image name
    ax.set_title(image_names[i])
    
    # Remove ticks from the plot.
    ax.set_xticks([])
    ax.set_yticks([])

    # Save to svg
    plt.savefig("images_visualization.png", bbox_inches="tight")
```
